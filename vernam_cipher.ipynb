```python
import string 

def vernam_encrypt(msg, key):
    cipher_text = ""
    pool = list(string.ascii_lowercase)

    for i in range(len(msg)):
        pt = pool.index(msg[i])
        k = pool.index(key[i])
        xor = pt ^ k
        
        if xor >= 26:
            xor = xor - 26
        
        cipher_text += pool[xor]

    return cipher_text


msg = input("Enter the message : ").lower()
key = input("Enter the key : ").lower()

if len(msg) == len(key):
    print("Encrypted text:", vernam_encrypt(msg, key))
else:
    print("Error! Message and key must be of same length.")
```
