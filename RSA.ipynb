```python
import random
import math

msg = input("Enter the message to encrypt: ").replace(" ", "")

def gen_primes(n1, n2):
    primes = []
    for num in range(n1, n2 + 1):
        if num > 1:
            for j in range(2, num):
                if num % j == 0:
                    break
            else:
                primes.append(num)
    return primes

def gen_PUPRkeys():
    p = random.choice(gen_primes(20, 50))
    q = random.choice(gen_primes(60, 90))
    n = p * q
    phi = (p - 1) * (q - 1)

    e = 2
    while e < phi:
        if math.gcd(e, phi) == 1:
            break
        e += 1

    i = 1
    while True:
        d = ((phi * i) + 1) / e
        if d == int(d):
            d = int(d)
            break
        i += 1

    return [e, d, n]

def RSA_encrypt(e, n, pt):
    codes = [ord(symbol) for symbol in pt]
    ct = [(code ** e) % n for code in codes]      # Fix 1 & 2: renamed + integer list
    return ct

def RSA_decrypt(d, n, ct):
    pt = [chr((code ** d) % n) for code in ct]    # Fix 1 & 3: renamed + decode ints
    return "".join(pt)

keys = gen_PUPRkeys()
e = keys[0]
d = keys[1]
n = keys[2]

ctxt = RSA_encrypt(e, n, msg)
print("Public key (e, n) :", e, ",", n)
print("Private key (d)   :", d)
print("Cipher Text       :", ctxt)
print("Decrypted Text    :", RSA_decrypt(d, n, ctxt))
```
