```python
import math

msg = input("Enter the message to encrypt: ").replace(" ", "")
key = input("Enter the key: ")

def gen_table(pt, col):
    rows = math.ceil(len(pt) / col)
    dcells = (rows * col) - len(pt)
    pt += 'x' * dcells

    table = []
    for i in range(0, len(pt), col):
        table.append(list(pt[i:i + col]))
    return table

def columnar(pt, keyword):
    ct = ""
    index = []
    matrix = gen_table(pt, len(keyword))
    sorted_keys = sorted(list(keyword))

    for char in sorted_keys:
        for i in range(len(keyword)):
            if keyword[i] == char and i not in index:
                index.append(i)
                for row in matrix:
                    ct += row[i]
                break
    return ct

ct = columnar(msg, key)
print("Cipher text:", ct)
```
