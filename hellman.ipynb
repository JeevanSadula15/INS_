```python
def calR1(x, p, q):
    R1 = (q ** x) % p
    return R1

def calR2(y, p, q):
    R2 = (q ** y) % p
    return R2

def calSK1(R2, x, p):
    sk1 = (R2 ** x) % p
    return sk1

def calSK2(R1, y, p):
    sk2 = (R1 ** y) % p
    return sk2


p = int(input("Enter p (prime number): "))
q = int(input("Enter q (primitive root / generator): "))
x = int(input("Enter x (private key of party1 / Alice): "))
y = int(input("Enter y (private key of party2 / Bob): "))

R1 = calR1(x, p, q)         
R2 = calR2(y, p, q)          

sk1 = calSK1(R2, x, p)       
sk2 = calSK2(R1, y, p)       

print("\nResults:")
print("Public value from Party1 (R1) =", R1)
print("Public value from Party2 (R2) =", R2)
print("Secret key at Party1         =", sk1)
print("Secret key at Party2         =", sk2)

if sk1 == sk2:
    print("\nKey exchange successful! Both parties have the same secret key.")
else:
    print("\nError: Keys do not match!")
```
