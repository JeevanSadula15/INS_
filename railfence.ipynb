```python
msg = input("Enter the message to encrypt: ").replace(" ", "")
rails = int(input("Enter the number of rails: "))

def railfence(msg, rails):
    if rails < 2:
        return msg  
    
    matrix = [['' for j in range(len(msg))] for i in range(rails)]
    
    row = 0
    check = 0  
    
    for char_index in range(len(msg)):
        matrix[row][char_index] = msg[char_index]
        
        if check == 0:
            row += 1
            if row == rails:
                check = 1
                row -= 2  
        else:
            row -= 1
            if row == 0:
                check = 0
                row = 1  
    
    ct = ""
    for i in range(rails):
        for j in range(len(msg)):
            if matrix[i][j] != '':
                ct += matrix[i][j]
    
    return ct

print("Encrypted Text:", railfence(msg, rails))
```
