```python
import string  

msg = input("Enter message to encrypt: ")

def encrypt(msg, key=3):
    pt = list(string.ascii_lowercase) + list(string.ascii_uppercase)
    cipher = ""
    for char in msg:           
        if char in pt:
            E = (pt.index(char) + key) % len(pt)
            cipher += pt[E]
        else:
            cipher += char     
    return cipher

print("Cipher text:", encrypt(msg))
```

```python
import string

cipher = input("Enter message to decrypt: ")

def decrypt(cipher, key=3):
    pt = list(string.ascii_lowercase) + list(string.ascii_uppercase)
    text = ""
    for char in cipher:          
        if char in pt:
            E = (pt.index(char) - key) % len(pt)
            text += pt[E]
        else:
            text += char        
    return text

print("Decrypted text:", decrypt(cipher))
```
