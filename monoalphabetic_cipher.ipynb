```python
import string
import random

msg = input("Enter message to encrypt: ")

def gen_key():
    symbols = list(string.ascii_lowercase)
    sym_shuff = list(string.ascii_lowercase)
    random.shuffle(sym_shuff)
    return dict(zip(symbols, sym_shuff))

key = gen_key()
print("Key: ", key)

def encrypt(msg, key):
    cipher = ""
    pt = msg.lower()
    for char in pt:
        if char in key.keys():
            cipher += key[char]
        else:
            cipher += char
    return cipher
    
cipher_txt = encrypt(msg, key)
print("Cipher text:", cipher_txt)


def decrypt(cipher, key):
    rev_key = dict(zip(key.values(), key.keys()))
    print("\nReversed Dictionary: ", rev_key)
    text = ""
    for char in cipher:
        if char in rev_key:
            text += rev_key[char]
        else:
            text += char
    return text

print("Decrypted text:", decrypt(cipher_txt, key))
```
